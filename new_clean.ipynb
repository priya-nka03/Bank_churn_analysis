```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_old=pd.read_csv("bank_chunkers.csv")
```

```python
df_old
```

```python
def exit_func(row):
    if row["Exited"]==True:
        return row
df_new=df_old.apply(exit_func,axis=1)
```

```python
df_new
```

```python
df=df_new.dropna()
```

```python
df
```
